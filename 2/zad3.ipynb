```python
import numpy as np

class neuralNetwork:
    data = []
    input_data=[]
    error = 0

    def __init__(self, first_matrix_size, weights_range):
        min_range = (weights_range[0] * 10)
        max_range = (weights_range[1] * 10)
        temp = np.random.randint(min_range, max_range, size=(first_matrix_size[0], first_matrix_size[1]))
        temp = temp.astype(float)
        temp = temp / 10
        self.data.append(temp)

    def add_layer(self, weights):
        if len(self.data) == 0:
            self.data.append(weights)

    def add_layer_rand(self, n, weights_range):
        if len(weights_range) < 2:
            return -1
        if weights_range[0] > weights_range[1]:
            return -2
        if weights_range[0] < -10 or weights_range[1] < -10:
            return -3
        if weights_range[0] > 10 or weights_range[1] > 10:
            return -4
        previous_shape = []
        previous_shape = self.data[len(self.data) - 1].shape
        min_range = (weights_range[0] * 10)
        max_range = (weights_range[1] * 10)

        temp = np.random.randint(min_range, max_range, size=(previous_shape[1], n))
        temp = temp.astype(float)
        temp = temp / 10
        # new_shape = temp.shape
        self.data.append(temp)

    def clear_data(self):
        self.data.clear()

    def predict(self, input_vector):
        return input_vector*self.data[0]


    def update_weights(self, weight_vector, expected_weight, alpha, cycles):        
        
        out2 = np.zeros((len(self.input_data), 4))
        out2 = np.asmatrix(out2)
        correct_counter=0

        for a in range(cycles):
            correct_counter=0
            for i in range(len(self.input_data)):
            
                # print(self.predict(np.asmatrix(self.input_data[i][0:3])))
                out2[i, :] = self.predict(np.asmatrix(self.input_data[i][0:3]))
                maxRow = out2[i,0]
                maxIndex = 0
                thisIterationCorrect = 0
                for j in range(4):
                    if out2[i,j]>maxRow:
                        maxRow=out2[i,j]
                        maxIndex = j
                if expected_weight[i][0,maxIndex]==1:
                    correct_counter+=1
        
                delta = (2*1 / int(self.data[0].shape[0])) * np.outer(
                    self.input_data[i][0:3],np.subtract(out2[i, :], expected_weight[i][:])
                )
                
                self.data[0] = self.data[0] - (delta * alpha)

        return correct_counter


    def smart_neural_network(self, weight_vector, expected_weight):
        out2 = np.zeros((len(self.input_data), 4))
        out2 = np.asmatrix(out2)
        correct_counter=0

        correct_counter=0
        for i in range(len(self.input_data)):
            out2[i, :] = self.predict(np.asmatrix(self.input_data[i][0:3]))
            maxRow = out2[i,0]
            maxIndex = 0
            thisIterationCorrect = 0
            for j in range(4):
                if out2[i,j]>maxRow:
                    maxRow=out2[i,j]
                    maxIndex = j
  
            if expected_weight[i][0,maxIndex]==1:
                correct_counter+=1
        return correct_counter


    def load_weights(self, file_name):
        np.savetxt('test1.txt', self.data[0])
        temp = np.loadtxt('test1.txt')
        print(temp)

    
    def load_multiple_weights(self, file_name):
        print("funload")
        temp = np.loadtxt(file_name)
        temp_matrix = np.zeros((4,3))
        temp_matrix = np.asmatrix(temp_matrix)
        how_many_matrixes = float(temp.shape[0]) / float(temp_matrix.shape[1])

        for i in range(int(how_many_matrixes)):
            temp_matrix = temp[(i*3):((i+1)*3),:]
            self.data.append(temp_matrix)
        

    def load_multiple_inputs(self,file_name):
        temp = np.loadtxt(file_name)
        temp_matrix = np.zeros((1,4))
        temp_matrix = np.asmatrix(temp_matrix)
        how_many_matrixes = float(temp.shape[0])
        # print(how_many_matrixes)
        for i in range(int(how_many_matrixes)):
            temp_matrix = temp[i,:]
            # print(temp_matrix)
            self.input_data.append(temp_matrix)

    def print_data(self):
        print(self.input_data)           


weights = np.matrix('0.1 0.1 -0.3;0.1 0.2 0.0;0.0 0.7 0.1;0.2 0.4 0.0')
alpha = 0.01
weights_range = [-0.6, 0.6]
matrix_size = [3, 4]
error_sum = 0

print("Wczytywanie pliku")
myNetwork2 = neuralNetwork(matrix_size, weights_range)
myNetwork2.load_multiple_inputs("training_colors.txt")
# przygotowanie expected_weights do tego zadania
# expected_weights=np.zeros((27*int(myNetwork2.input_data[0].shape[0]),1))
expected_weights=[]
# expected_weights = np.asmatrix(expected_weights)
one = np.zeros((1,4))
two = np.zeros((1,4))
three = np.zeros((1,4))
four = np.zeros((1,4))
one=np.asmatrix(one)
two = np.asmatrix(two)
three = np.asmatrix(three)
four = np.asmatrix(four)
one[0, 0] = 1
two[0, 1] = 1
three[0, 2] = 1
four[0, 3] = 1

for i in range(len(myNetwork2.input_data)):
    # print(myNetwork2.input_data[i])
    # print(int(myNetwork2.input_data[i][-1]))
    if myNetwork2.input_data[i][-1]==1:
        expected_weights.append(one)
    elif myNetwork2.input_data[i][-1]==2:
        expected_weights.append(two)
    elif myNetwork2.input_data[i][-1]==3:
        expected_weights.append(three)
    elif myNetwork2.input_data[i][-1]==4:
        expected_weights.append(four)


correct=myNetwork2.update_weights(weights,expected_weights,alpha,15)
print(correct)
print("Procent poprawnych")
print(correct/109*100)

#teraz dane testowe
print("DANE TESTOWE:")
myNetwork2.input_data.clear()
myNetwork2.load_multiple_inputs("test_colors.txt")
# print(len(myNetwork2.input_data))
expected_weights.clear()
for i in range(len(myNetwork2.input_data)):
    # print(myNetwork2.input_data[i])
    # print(int(myNetwork2.input_data[i][-1]))
    if myNetwork2.input_data[i][-1]==1:
        expected_weights.append(one)
    elif myNetwork2.input_data[i][-1]==2:
        expected_weights.append(two)
    elif myNetwork2.input_data[i][-1]==3:
        expected_weights.append(three)
    elif myNetwork2.input_data[i][-1]==4:
        expected_weights.append(four)

correct = myNetwork2.smart_neural_network(weights,expected_weights)
print(correct)
print("Procent poprawnych")
print(correct/130*100)


```
